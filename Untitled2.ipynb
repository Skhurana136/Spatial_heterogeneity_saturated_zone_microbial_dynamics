```python
#Load standard values required to navigate through the datasets
data_dir = "Y:/Home/khurana/4. Publications/Restructuring/Paper1/Figurecodes/"
filename = "biomass_comparison_Original_complete_06022021.csv"
Regimes = ["Slow", "Medium", "Fast"]
vels = [0.00038,0.0038,0.038]
gw = 1

scdict = proc.masterscenarios() #master dictionary of all spatially heterogeneous scenarios that were run
ratenames = proc.masterrates("saturated")

#Domains
Trial = list(t for t,values in scdict.items())
#Het = list(values['Het'] for t,values in scdict.items())
#Anis = list(values['Anis'] for t,values in scdict.items())

#Reactive species of concerns
States = ["Active", "Inactive"]
Locations = ["Mobile", "Immobile"]
allspecies = proc.speciesdict("Saturated")#gvarnames = ["DOC","DO","Ammonium","Nitrate","Nitrogen","TOC"]
microbialspecies = list(t for t in allspecies.keys() if allspecies[t]["State"] in States)
```

```python
#chem_path_data = r"Y:\Home\khurana\4. Publications\Restructuring\Paper1\Figurecodes\massflux_withbreakthrough_forMartin_v4_complete.csv"
biomass_data_path = os.path.join(data_dir, filename)
allbiomassdata = pd.read_csv(biomass_data_path, sep = '\t')
print(allbiomassdata.columns)
allbiomassdata['Regime'] = allbiomassdata['Regime'].replace({'Equal':'Medium'})
biomassdata = allbiomassdata[allbiomassdata.Time_series==0]
```

```python
print(biomassdata.Time_series.unique())
print(biomassdata.Time_series.unique())
print(biomassdata.Regime.unique())
uniquespecies = biomassdata.Chem.unique().tolist()
print(biomassdata.Regime)
```

# Assign each row in the dataset with a state and location. This is easier to summarize in the paper

```python
for s in uniquespecies:
    if s == "Total":
        biomassdata.loc[biomassdata.Chem == s, 'State'] = 'NA'
        biomassdata.loc[biomassdata.Chem == s, 'Location'] = 'NA'
    else:
        biomassdata.loc[biomassdata.Chem == s, 'State'] = allspecies[s]["State"]
        biomassdata.loc[biomassdata.Chem == s, 'Location'] = allspecies[s]['Location']
```

```python
col_dit = {'Slow':'indianred', 'Medium':'g', 'Fast':'steelblue'}
uniquespecies.remove('Total')
uniquespecies.remove('Immobile active sulphate reducers')
uniquespecies.remove('Immobile inactive sulphate reducers')
uniquespecies.remove('Mobile active sulphate reducers')
uniquespecies.remove('Mobile inactive sulphate reducers')
print(uniquespecies)
```

```python
fig, axeses = plt.subplots(4,3, sharex = True, figsize = (15,15))
for c in uniquespecies:
    subdata = biomassdata[biomassdata.Chem ==c]
    ax = axeses.flat[uniquespecies.index(c)]
    ax.set_title(c)
    for r in ["Slow", "Medium", "Fast"]:
        data = subdata[subdata.Regime == r]
        ax.scatter(data.fraction, 100*data.Contribution, c = col_dit[r])
plt.savefig(os.path.join(data_dir, "contribution_microbial_biomass.png"), dpi = 300)
```

```python
x = biomassdata.groupby(['Regime','Trial','State', 'Location'], as_index=False)['Contribution'].sum()
```

```python
x = pd.merge(x, biomassdata[['Regime', 'Trial', 'Time', 'fraction']], on = ['Regime', 'Trial'])
x.to_csv(os.path.join(data_dir, "x.csv"))
```

```python
fig, axeses = plt.subplots(2,2, sharex = True, sharey = True, figsize = (10,10))
i = 0
for a in ["Active", "Inactive"]:
    for l in ["Immobile", "Mobile"]:
        subdata = x[(x.State == a) & (x.Location == l)]
        ax = axeses.flat[i]
        ax.set_title(a + " " + l + " biomass", fontsize = 18)
        for r in ["Slow", "Medium", "Fast"]:
            data = subdata[subdata.Regime == r]
            ax.scatter(100*data.fraction, 100*data.Contribution, c = col_dit[r], label = r + " flow")
        i += 1
ax.legend(fontsize = 16)
axeses.flat[2].set_xlabel ("Residence time of solutes (%)", fontsize = 16)
axeses.flat[3].set_xlabel ("Residence time of solutes (%)", fontsize = 16)
axeses.flat[0].set_ylabel ("Contribution to total biomass (%)", fontsize = 16)
axeses.flat[2].set_ylabel ("Contribution to total biomass (%)", fontsize = 16)
plt.savefig(os.path.join(data_dir, "contribution_state_location_biomass.png"), dpi = 300)
```
